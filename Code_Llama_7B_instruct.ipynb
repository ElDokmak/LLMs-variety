```python
import transformers
from transformers import AutoTokenizer
import torch
import accelerate
```

```python
model = "codellama/CodeLlama-7b-Instruct-hf"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
system = "Provide answers in Python"
user = "Write a function that detects a pattern that matches the style of 23-Jan-2023 from the given text"

prompt = f"<s><<SYS>>\n{system}\n<</SYS>>\n\n{user}"
# Note the the previos line is the prompt of instrct llama
# if u are using anthor version u should use its prompt
```

```python
sequences = pipeline(
    prompt,
    do_sample=True,
    top_k=10,
    temperature=0.1,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=200,
    add_special_tokens=False
)
```

```python
for seq in sequences :
  print(f"Result : {seq['generated_text']}")
```

```python
# Let's try the result and check the answer
import re

def detect_pattern(text):
  pattern = r"\d{2}/\d{2}/\d{4}"
  return re.findall(pattern, text)
```
